```python
seoul = pd.read_csv("../data/코로나이전명지_final.csv")
del seoul['Unnamed: 0']
seoul.columns=['recommend','comment','date','title','content']
seoul['title'] = seoul['title'].str.replace("[^ㄱ-하-ㅣ가-힣]"," ") # 한글이 아니면 전부 제거
seoul['content'] = seoul['content'].str.replace("[^ㄱ-하-ㅣ가-힣]"," ") # 한글이 아니면 전부 제거

import matplotlib.pyplot as plt
import urllib.request
from gensim.models.word2vec import Word2Vec
from konlpy.tag import Okt
import numpy as np

okt = Okt()
train_data = seoul
train_data[:30] # 상위 5개 출력

train_data = train_data.dropna()

word_list = []

for i in train_data.title:
    word_list.append(i)
for i in train_data.content:
    word_list.append(i)
    
sentences_tag = []

for sentence in word_list:
    morph = okt.pos(sentence)
    sentences_tag.append(morph)
    
# 명사, 동사, 형용사 사용

noun_list = []
for sentence in sentences_tag:
    for word, tag in sentence:
        if tag in ["Noun","Verb", "Adjective"]:
            noun_list.append(word)
            
len(noun_list)

noun_list = [n for n in noun_list if len(n) > 1]
noun_list[:100]


from collections import Counter
from wordcloud import WordCloud

counts = Counter(noun_list)
tags = counts.most_common(100)
tags

tags = dict(tags)

tags.pop('학기')
tags.pop('언제')
tags.pop('오늘')
tags.pop('진짜')
tags.pop('어떻게')
tags.pop('하는')
tags.pop('하나')
tags.pop('한정')
tags.pop('이번')
tags.pop('해서')
tags.pop('그냥')
tags.pop('생각')
tags.pop('했는데')
tags.pop('정도')
tags.pop('확인')
tags.pop('있나요')
tags.pop('학년')
tags.pop('떴다')
tags.pop('혹시')
tags.pop('지금')
tags.pop('하면')
tags.pop('하는데')
tags.pop('있는')
tags.pop('내일')
tags.pop('이제')
tags.pop('건가')
tags.pop('평균')
tags.pop('시간')
tags.pop('우리')
tags.pop('여자')
tags.pop('아시')
tags.pop('기윤')
tags.pop('어디')
tags.pop('신분')
tags.pop('남자')
tags.pop('해야')
tags.pop('보고')
tags.pop('메일')
tags.pop('입력')
tags.pop('하고')
tags.pop('나옴')
tags.pop('들어')
tags.pop('이해')
tags.pop('같은데')
tags.pop('원래')
tags.pop('한다')
tags.pop('하기')
tags.pop('한번')
tags.pop('아님')
tags.pop('동안')
```

# 코로나 이전 이후 비교

```python
plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
sns.barplot(data=seoul_before_corona, x='count', y='word',color=sns.color_palette("Paired")[0])
plt.title("코로나 이전 명지")
plt.subplot(1,2,2)
sns.barplot(data=yongin_before_corona, x='count', y='word',color=sns.color_palette("Paired")[4])
plt.title("포스트 코로나 명지")
plt.show()
```

- 코로나 이전: 성적, 시험 등 학업에 관련한 단어가 많이 출현한것으로 보여짐
- 코로나 이후: 학교라는 단어의 출현 빈도가 상승하였고 비대면의 장기화로 장학금(등록금)과 관련한 단어가 많이 출현한것으로 보여짐
- 또한 사람, 대면, 친구, 학우 등의 인간관계와 관련된 단어가 많이 출현하였다.
- 코로나 이전과 비교했을 때 보이지 않던 휴학 이라는 단어가 상위 빈출 단어로 출현한 것으로 분석되었다.

```python
plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
sns.barplot(data=merge_before_corona, x='count_y', y='word',color=sns.color_palette("Paired")[0])
plt.title("코로나 이전 자연캠퍼스")
plt.subplot(1,2,2)
sns.barplot(data=merge_after_corona, x='count_y', y='word',color=sns.color_palette("Paired")[4])
plt.title("포스트 코로나 자연캠퍼스")
plt.show()

plt.savefig('../docs/yongin_barplt.png')
```

- 자연캠퍼스도 마찬가지로 학교, 장학금 등의 단어가 가장 많이 출현하였다.
- 또한 장학금(등록금)의 단어가 많이 출현하였으며 대면, 사람, 학우 등의 단어가 나타났다.
- 자연 캠퍼스 역시도 코로나 이전에 보이지 않던 휴학과 대인관계, 장학금과 관련한 단어가 많이 나타난 것으로 분석된다.
